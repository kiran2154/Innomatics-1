```python
def analyze_engagement(likes_list):
  total_likes=0
  for likes in likes_list:
    total_likes+=likes
  print("Your Total Likes:",total_likes)
  if total_likes >= 1000:
    print("Viral Post")
  else:
    print("Normal Engagement")
likes=[]
n = int(input("Enter number of like entries: "))
for i in range(n):
    value = int(input(f"Enter likes for entry {i+1}: "))
    likes.append(value)
analyze_engagement(likes)
```

```python
def analyze_rainfall(data,required_level):
  total=0
  for rain in data:
    total+=rain
  avg=total/len(data)
  print("Average Rainfall in this Season:",int(avg))
  if avg >= required_level:
    print("Adequate Rainfall")
  else:
    print("Inadequate Rainfall")
rainfall = [60, 80, 75, 70, 75]   #average=72
required = 70
analyze_rainfall(rainfall, required)
```

```python
def duplicate_username(usernames):
  unique_usernames = set(usernames)
  if len(usernames) != len(unique_usernames):
    print("Duplicate Accounts Found: Yes")
  else:
    print("Duplicate Accounts Found: No")
# Example Input
usernames = ["user1", "user2", "user3", "user1"]

duplicate_username(usernames)
```

```python
def filter_crop_price(prices):
    premium_crops = []
    for price in prices:
        if price > 2000:
            premium_crops.append(price)
    print("Premium Crops:", premium_crops)
# Example Input
crop_prices = [1500, 2500, 1800, 3200, 1900]
filter_crop_price(crop_prices)
```

```python
def eval_application_health(error_count):
  print("Error Count: ",error_count)
  if error_count == 0:
    status = "Healthy"
  elif error_count <= 5:
    status = "Minor Issues"
  else:
    status = "Critical Issues"
  print("System Status: ",status)
eval_application_health(7)
```

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    # Calculate percentage
    attendance_percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", attendance_percentage)
    # Check eligibility
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
# Example Input
attendance = ["P", "P", "A", "P", "P"]  #80%
check_exam_eligibility(attendance)
```

```python
def calculate_electricity_bill(units):
    bill = 0
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
# Example Usage
units_consumed = 400
total_bill, usage_status = calculate_electricity_bill(units_consumed)
print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
