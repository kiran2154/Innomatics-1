```python
marks = [45, 78, 90, 33, 60]
passed = []
failed = []
for mark in marks:
  if mark >= 50:
    passed.append(mark)
  else:
    failed.append(mark)
print("Passed Marks:", passed)
print("Failed Marks:", failed)
print("Total Passed:", len(passed))
print("Total Failed:", len(failed))
```

```python
names = [" Alice ", "bob", " CHARLIE "]
clean_List =  []
for names in names:
  clean_List.append(names.strip().lower())
print("Cleaned names: ",clean_List)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
  length=len(message)
  print("Message: ",message)
  print("Length: ",length)
  if length > 10:
    print("This message is longer than 10 characters")
```
