```python
def get_unique_words(sentence):
    words = sentence.lower().split()
    unique_words = list(dict.fromkeys(words))
    return unique_words
sentence = input("Enter Your Sentence: ")
result = get_unique_words(sentence)

print("Unique Words:", result)
print("Count:", len(result))
```

```python
numbers = [45, 22, 89, 10, 66]
#intialization
maximum = numbers[0]
minimum = numbers[0]
#loop
for number in numbers:
    if number > maximum:
        maximum = number
    elif number < minimum:
        minimum = number
#output
print("List:",numbers)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
#given list of price
prices = [450, 1200, 899, 1500, 300]
count=0
#loops to compare each price
for price in prices:
    if price > 1000:
        count += 1
print("Products above 1000: ",count)
```

```python
#given list of attendance
attendance = ["P", "P", "A", "P", "P"]
present=0
for status in attendance:
    if status == "P":
        present += 1
#total days
total_days = len(attendance)
attendance_percentage = (present / total_days) * 100
#output
print("Attendance Percentage:", attendance_percentage)
```

```python
#assume text as pythonp based on the expected outcome
text = "pythonp"
#character frequency
char_count = {}
#loop
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
#outupt stored in char_count
print(char_count)
```
