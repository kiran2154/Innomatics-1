```python
employees = { "Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}
#highest score
highest_score=max(employees.values())
#top performance
top_performers=[]
#to handle ties
for name,score in employees.items():
    if score==highest_score:
        top_performers.append(name)
print("Top performers eligible for Bonus: ",", ".join(top_performers),f"(Score: {highest_score})")
```

```python
import string
word = "Buy mobile phone buy phone online"
#Convert to lowercase
word =word.lower()
#Remove punctuation
for char in string.punctuation:
  word = word.replace(char, " ")
#Split into words
words = word.split()
#Count frequency
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
repeated_keywords = {}
for word, count in frequency.items():
    if count > 1:
        repeated_keywords[word] = count
print(repeated_keywords)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []
for index, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if even
        valid_readings.append((index, value))
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_counts = {}
#Extract domain
for email in emails:
    domain = email.split("@")[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1
#percentage
total_users = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
#average sales
average=sum(sales)/len(sales)
#sales above 30%
dis=average * 1.3
dis = round(dis, -2)
#spike day
for day, value in enumerate(sales, start=1):
    if value >= dis:
        print(f"Day {day}: {value}")
```
