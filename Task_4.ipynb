```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Available plans are:", valid_plans)
        return False
    else:
        return True
while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        if validate_recharge(amount):
            print("Recharge Successful! ")
            break
        else:
            print("Please try again.\n")
    except ValueError:
        print("Invalid input! Please enter a valid number.\n")
```

```python
def check_stock(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(f"{product} : Reorder Alert ")
        else:
            print(f"{product} : Stock OK ")
#Dictionary
inventory = {
    "Mobile": 25,
    "Laptop": 10,
    "Computer": 25,
    "Headphones": 8
}
check_stock(inventory)
```

```python
def marks():
    marklist=list(map(int,input("Enter marks:").split(',')))
    total=0
    for i in marklist:
        total+=i
    avg=total/len(marklist)
    if avg>=50:
        print("Pass")
    else:
        print("Fail")
marks()
```

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    if peak == "yes":
        fare = fare + (fare * 0.25)
    return fare
while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Peak hour? (yes/no): ").lower()
    total = calculate_fare(distance, peak)
    print("Total Fare: ₹", total)
    retry = input("Try again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
def check_attendance(attendance):
    total_days = len(attendance)
    present_days = 0
    for day in attendance:
        if day.upper() == "P":
            present_days += 1
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
record = input("Enter attendance records (e.g., PPPAP): ")
print(check_attendance(record))
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    if len(password) < 8:
        return "Weak Password (Minimum 8 characters required)"
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if has_digit and has_special:
        return "Strong Password "
    else:
        return "Weak Password "
# Example
pswd = input("Enter password: ")
result = check_password(pswd)
print(result)
```
